```python
#importing required libraries

import pandas as pd

from selenium import webdriver

from selenium.common.exceptions import NoSuchElementException

import time
```

```python
#Connecting with chrome driver

driver = webdriver.Chrome("Chromedriver.exe")
```

```python
#opening Amazon URL in chrome

driver.get('https://www.amazon.in/')
```

```python
#Sending required input in search bar

search_tab = driver.find_element_by_id('twotabsearchtextbox')
search_tab.send_keys('Shoes')
```

```python
#Clicking on Search Button to open input result

search_button = driver.find_element_by_id('nav-search-submit-button')
search_button.click()
```

```python
#Collecting URL's of the result products in the first three pages

URL = []

for i in range(0,3):
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        URL.append(j.get_attribute('href'))
    driver.find_element_by_xpath("//ul[@class='a-pagination']/li[7]").click()
    time.sleep(4)
```

```python
print(len(URL)) #To check the count/values
```

```python
#Creating empty list for all the required variables

Brand_Name = []
Name_of_the_Product = []
Rating = []
Number_of_Ratings = []
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
Other_Details = []
```

```python
#Applying for loop for the URL's to get required data

for i in URL:
    driver.get(i)
    
    Brand_tag = driver.find_element_by_id('bylineInfo')
    Brand_Name.append(Brand_tag.text)
    
    Product_title = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
    Name_of_the_Product.append(Product_title.text)
    
    try:
        rating = driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append("No Rating")
    
    try:
        Rating_tag = driver.find_element_by_xpath("//div[@class='a-row a-spacing-medium averageStarRatingNumerical']")
        Number_of_Ratings.append(Rating_tag.text)
    except NoSuchElementException:
        Number_of_Ratings.append("No Ratings")
    
    try:
        price_tag = driver.find_element_by_xpath("//td[@class='a-span12']")
        Price.append(price_tag.text)
    except NoSuchElementException:
        Price.append("-")
    
    try:
        Return_tag = driver.find_element_by_id('RETURNS_POLICY')
        Return_Exchange.append(Return_tag.text)
    except NoSuchElementException:
        Return_Exchange.append("No Return")
    
    try:
        Delivery_tag = driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")
        Expected_Delivery.append(Delivery_tag.text)
    except NoSuchElementException:
        Expected_Delivery.append("-")
    
    try:
        Availability_tag = driver.find_element_by_id('availability')
        Availability.append(Availability_tag.text)
    except NoSuchElementException:
        Availability.append("-")
    
    other_details_tag = driver.find_element_by_id('feature-bullets')
    Other_Details.append(other_details_tag.text.replace('\n',' '))
```

```python
print(len(Brand_Name), len(Name_of_the_Product), len(Rating), len(Number_of_Ratings), len(Price), len(Return_Exchange), len(Expected_Delivery), len(Availability), len(Other_Details))
```

```python
#Creating DataFrame with the results

Shoes = pd.DataFrame({})

Shoes['Brand Name'] = Brand_Name
Shoes['Title'] = Name_of_the_Product
Shoes['Price'] = Price
Shoes['Availability'] = Availability
Shoes['Expected Delivery'] = Expected_Delivery
Shoes['Return/Exchange'] = Return_Exchange
Shoes['Rating'] = Rating
Shoes['Number of Ratings'] = Number_of_Ratings
Shoes['Other Details'] = Other_Details

Shoes
```

```python
#Saving to CSV

import csv

Shoes.to_csv('Amazon_Shoes.csv', quoting=csv.QUOTE_NONNUMERIC)
```

```python
#Connecting with chrome driver

driver = webdriver.Chrome('chromedriver.exe')
```

```python
#opening Flipkart URL in chrome

driver.get('http://www.flipkart.com/')
```

```python
#Closing pop-ups if any

pop_up = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
pop_up.click()
```

```python
#Sending required input in search bar

search_bar = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_bar.send_keys('Smart Phones')
```

```python
#Clicking on Search Button to open input result

search_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
#Collecting URL's of the result products in the first page

url = []
product_url = driver.find_elements_by_xpath("//a[@class='_1fQZEK']")

for i in product_url:
    url.append(i.get_attribute('href'))
    
url
```

```python
#Creating empty list for all the required variables

Brand_Name = []
Smartphone_Name = []
Colour = []
RAM = []
Storage = []
Primary_Camera = []
Secondary_Camera = []
Display_Size = []
Display_Resolution = []
Processor = []
Processor_Cores = []
Battery_Capacity = []
Price = []
```

```python
#Applying for loop for the URL's to get required data

for i in url:
    driver.get(i)
    
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    
    title = driver.find_element_by_xpath("//span[@class='B_NuCI']")
    Smartphone_Name.append(title.text)
    
    name = driver.find_element_by_xpath("//div[@class='_1MR4o5']/div[4]/a")
    x = name.text.replace('Mobiles','').strip()
    Brand_Name.append(x)
    
    Prices = driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
    Price.append(Prices.text)
    
    color = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')
    Colour.append(color.text)
    
    Display_Sizes = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table/tbody/tr[1]/td[2]/ul/li")
    Display_Size.append(Display_Sizes.text)
    
    Resolution = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table/tbody/tr[2]/td[2]/ul/li")
    Display_Resolution.append(Resolution.text)
    
    Processors = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table/tbody/tr[2]/td[2]/ul/li")
    Processor.append(Processors.text)
    
    Processor_Core = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table/tbody/tr[3]/td[2]/ul/li")
    Processor_Cores.append(Processor_Core.text)
    
    ROM = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table/tbody/tr[1]/td[2]/ul/li")
    Storage.append(ROM.text)
    
    RAM_tag = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table/tbody/tr[2]/td[2]/ul/li")
    RAM.append(RAM_tag.text)
    
    P_Camera = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table/tbody/tr[2]/td[2]/ul/li")
    Primary_Camera.append(P_Camera.text)
    
    S_Camera = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table/tbody/tr[5]/td[2]/ul/li")
    Secondary_Camera.append(S_Camera.text)
    
    Battery = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][10]/table/tbody/tr[1]/td[2]/ul/li")
    Battery_Capacity.append(Battery.text)
    
    time.sleep(4)
```

```python
Brand_Name
```

```python
print(len(Brand_Name), len(Smartphone_Name), len(Colour), len(RAM), len(Storage), len(Primary_Camera), len(Secondary_Camera), len(Display_Size), len(Display_Resolution), len(Processor), len(Processor_Cores), len(Battery_Capacity), len(Price))
```

```python
#Creating DataFrame with the results

Smartphones = pd.DataFrame({})

Smartphones['Brand Name'] = Brand_Name
Smartphones['Title'] = Smartphone_Name
Smartphones['Color'] = Colour
Smartphones['Storage'] = Storage
Smartphones['RAM'] = RAM
Smartphones['Primary Camera'] = Primary_Camera
Smartphones['Secondary Camera'] = Secondary_Camera
Smartphones['Display Size'] = Display_Size
Smartphones['Display Resolution'] = Display_Resolution
Smartphones['Processor'] = Processor
Smartphones['Processor Cores'] = Processor_Cores
Smartphones['Battery Capacity'] = Battery_Capacity
Smartphones['Price'] = Price

Smartphones
```

```python
#Saving to CSV

import csv

Smartphones.to_csv('Smartphones.csv', quoting=csv.QUOTE_NONNUMERIC)
```
