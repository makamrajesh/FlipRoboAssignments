# Scraping Header Tags from wikipedia

```python
#importing the required libraries

from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
print(soup.prettify())
```

```python
#First we will extract the html tag where we have the first Heading

first_Heading = soup.find('h2', class_='mp-h2')
first_Heading
```

```python
#print first Heading

first_Heading.text
```

```python
#we will extract the all html tag where we have Headings

Headings = soup.find_all('h2', class_='mp-h2')
Headings
```

```python
#Lisitng all the headers

wiki_Headings = []

for i in Headings:
    wiki_Headings.append(i.text.replace('\n','').strip())

wiki_Headings
```
