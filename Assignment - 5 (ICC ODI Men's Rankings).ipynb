# Scraping Data of ODI Ranks and Player Ranks using Beautifulsoup from ICC-Cricket Website

```python
#importing the required libraries

from bs4 import BeautifulSoup
import requests
```

```python
#ODI Rankings

page1 = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page1
```

```python
soup = BeautifulSoup(page1.content)
print(soup.prettify())
```

```python
ODI_Teams1 = soup.find_all('td', class_='rankings-block__banner--team-name')
ODI_Teams1
```

```python
ODI_Teams_list = []

for i in ODI_Teams1:
     for j in i.find_all("span", class_= 'u-hide-phablet'):
            ODI_Teams_list.append(j.text)

ODI_Teams_list
```

```python
ODI_Teams2 = soup.find_all('td', class_='table-body__cell rankings-table__team')
ODI_Teams2
```

```python
for i in ODI_Teams2[0:9]:
     for j in i.find_all("span", class_= 'u-hide-phablet'):
            ODI_Teams_list.append(j.text)

ODI_Teams_list
```

```python
Matches1 = soup.find_all('td' ,class_="rankings-block__banner--matches")
Matches1
```

```python
ODI_Matches_list = []

for i in Matches1:
     ODI_Matches_list.append(i.text)

ODI_Matches_list
```

```python
Points1 = soup.find_all('td' ,class_="rankings-block__banner--points")
Points1
```

```python
ODI_Points_list = []

for i in Points1:
     ODI_Points_list.append(i.text)

ODI_Points_list
```

```python
Matches2 = soup.find_all('td', class_="table-body__cell u-center-text")
Matches2
```

```python
R_l = []

for i in Matches2[0:18]:
    R_l.append(i.text.replace('\n','').strip())
R_l

for i in range(len(R_l)):
    if( i%2 == 0):
        ODI_Matches_list.append(R_l[i])
    else:
        ODI_Points_list.append(R_l[i])
```

```python
ODI_Matches_list 
```

```python
ODI_Points_list
```

```python
Ratings1 = soup.find_all('td', class_="rankings-block__banner--rating u-text-right")
Ratings1
```

```python
Ratings2 = soup.find_all('td', class_="table-body__cell u-text-right rating")
Ratings2
```

```python
ODI_Ratings = []

for i in Ratings1:
    ODI_Ratings.append(i.text.replace('\n','').strip())

for i in Ratings2[0:9]:
    ODI_Ratings.append(i.text.replace('\n','').strip())

ODI_Ratings
```

```python
print(len(ODI_Teams_list), len(ODI_Matches_list), len(ODI_Points_list), len(ODI_Ratings))
```

```python
import pandas as pd

ODI_Mens_Cricket = pd.DataFrame({})

ODI_Mens_Cricket['Team'] = ODI_Teams_list
ODI_Mens_Cricket['Matches'] = ODI_Matches_list
ODI_Mens_Cricket['Points'] = ODI_Points_list 
ODI_Mens_Cricket['Rating'] = ODI_Ratings

ODI_Mens_Cricket
```

```python
#Top 10 ODI Batsmen

page2 = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
print(soup2.prettify())
```

```python
Players = soup2.find_all('div', class_="rankings-block__banner--name-large")
Players
```

```python
Players2 = soup2.find_all('td', class_='table-body__cell rankings-table__name name')
Players2
```

```python
Players_list = []

for i in Players:
     Players_list.append(i.text)
        
for i in Players2[0:9]:
     for j in i.find_all("a"):
            Players_list.append(j.text)

Players_list
```

```python
Team1 = soup2.find_all('div', class_="rankings-block__banner--nationality")
Team1
```

```python
Team2 = soup2.find_all('span', class_="table-body__logo-text")
Team2
```

```python
Teams_list = []

for i in Team1:
     Teams_list.append(i.text.replace('\n','').strip())
        
for i in Team2[0:9]:
    Teams_list.append(i.text.replace('\n','').strip())

Teams_list
```

```python
Rating1 = soup2.find_all('div', class_="rankings-block__banner--rating")
Rating1
```

```python
Rating2 = soup2.find_all('td', class_="table-body__cell rating")
Rating2
```

```python
Ratings_list = []

for i in Rating1:
     Ratings_list.append(i.text.replace('\n','').strip())
        
for i in Rating2[0:9]:
    Ratings_list.append(i.text.replace('\n','').strip())

Ratings_list
```

```python
Record1 = soup2.find_all('span', class_="rankings-block__career-best-text")
Record1
```

```python
Record2 = soup2.find_all('td', class_="table-body__cell u-text-right u-hide-phablet")
Record2
```

```python
Records_list = []

for i in Record1:
     Records_list.append(i.text.replace('\n','').strip())
        
for i in Record2[0:9]:
    Records_list.append(i.text.replace('\n','').strip())

Records_list
```

```python
print(len(Players_list), len(Teams_list), len(Ratings_list), len(Records_list))
```

```python
import pandas as pd

ODI_Batting_Men = pd.DataFrame({})

ODI_Batting_Men['Player'] = Players_list
ODI_Batting_Men['Team'] = Teams_list
ODI_Batting_Men['Rating'] = Ratings_list 
ODI_Batting_Men['Career_Best_Rating'] = Records_list

ODI_Batting_Men
```
