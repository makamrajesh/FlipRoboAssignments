```python
import pandas as pd

from selenium import webdriver

import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.myntra.com/shoes')
```

```python
color_tag = driver.find_elements_by_xpath("//span[@class='colour-label colour-colorDisplay']")

for i in color_tag:
    if i.text == 'Black':
        i.click()
        break
```

```python
price_tag = driver.find_elements_by_xpath("//ul[@class='price-list']/label")

for i in price_tag:
    if i.text == 'Rs. 6649 to Rs. 13099':
        i.click()
        break
```

```python
Brand = []
Shoe_Description = []
Price = []
```

```python
for i in range(0,2):
    
    for j in driver.find_elements_by_xpath("//h3[@class='product-brand']"):
        Brand.append(j.text)
        
    for k in driver.find_elements_by_xpath("//h4[@class='product-product']"):
        Shoe_Description.append(k.text.replace('\n',' '))
     
    for m in driver.find_elements_by_xpath("//div[@class='product-price']"):
        v=m.text.strip().split('Rs.')[1].strip()
        print(v)
        Price.append(v)
            
    driver.find_element_by_xpath("//li[@class='pagination-next']").click()
    
    time.sleep(1)
```

```python
print(len(Brand),len(Shoe_Description),len(Price))
```

```python
Shoes = pd.DataFrame({})

Shoes['Brand'] = Brand
Shoes['Description'] = Shoe_Description
Shoes['Price'] = Price

Shoes
```
