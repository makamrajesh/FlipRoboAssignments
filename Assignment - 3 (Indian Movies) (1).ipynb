```python
# Displaying IMDBâ€™s Top rated 100 Indian movies data by web scraping using beautifulsoup library
```

```python
#importing the required libraries

from bs4 import BeautifulSoup
import requests
```

```python
Indian_Movies = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
Indian_Movies
```

```python
Indian_Movies.content
```

```python
soup1=BeautifulSoup(Indian_Movies.content)
print(soup1.prettify())
```

```python
Movies1_100 = soup1.find_all('td', class_='titleColumn')
Movies1_100
```

```python
import re

Movies100 = []

for i in Movies1_100[0:100]:

  value = i.text
  name = value.split('\n') # it will split the value based on new line (\n)
  value = name[2]
  val = re.sub('\s\s+', '',value) # it will remove 2 or more that 2 spaces
  
  Movies100.append(val)

Movies100
```

```python
Rating1_100 = soup1.find_all('td', class_='ratingColumn imdbRating')
Rating1_100
```

```python
Rating100 = []

for i in Rating1_100[0:100]:
    Rating100.append(i.text.strip())

Rating100
```

```python
Year1_100 = soup1.find_all('span', class_='secondaryInfo')
Year1_100
```

```python
Year100 = []

for i in Year1_100[0:100]:
    Year100.append(i.text.strip())

Year100
```

```python
print(len(Movies100), len(Year100), len(Rating100))
```

```python
import pandas as pd

Movies = pd.DataFrame({})

Movies['Name'] = Movies100
Movies['Year'] = Year100
Movies['Rating'] = Rating100

Movies
```

```python
import csv

Movies.to_csv('Indian_Movies.csv', index=False, quoting=csv.QUOTE_NONNUMERIC)
```
