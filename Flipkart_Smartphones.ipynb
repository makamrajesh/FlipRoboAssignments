```python
import pandas as pd

from selenium import webdriver

from selenium.common.exceptions import NoSuchElementException

import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('http://www.flipkart.com/')
```

```python
pop_up = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
pop_up.click()
```

```python
search_bar = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_bar.send_keys('Smart Phones')
```

```python
search_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
url = []
product_url = driver.find_elements_by_xpath("//a[@class='_1fQZEK']")

for i in product_url:
    url.append(i.get_attribute('href'))
    
url
```

```python
Brand_Name = []
Smartphone_Name = []
Colour = []
RAM = []
Storage = []
Primary_Camera = []
Secondary_Camera = []
Display_Size = []
Display_Resolution = []
Processor = []
Processor_Cores = []
Battery_Capacity = []
Price = []
```

```python
for i in url:
    driver.get(i)
    
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    
    title = driver.find_element_by_xpath("//span[@class='B_NuCI']")
    Smartphone_Name.append(title.text)
    
    name = driver.find_element_by_xpath("//div[@class='_1MR4o5']/div[4]/a")
    x = name.text.replace('Mobiles','').strip()
    Brand_Name.append(x)
    
    Prices = driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
    Price.append(Prices.text)
    
    color = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')
    Colour.append(color.text)
    
    Display_Sizes = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table/tbody/tr[1]/td[2]/ul/li")
    Display_Size.append(Display_Sizes.text)
    
    Resolution = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table/tbody/tr[2]/td[2]/ul/li")
    Display_Resolution.append(Resolution.text)
    
    Processors = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table/tbody/tr[2]/td[2]/ul/li")
    Processor.append(Processors.text)
    
    Processor_Core = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table/tbody/tr[3]/td[2]/ul/li")
    Processor_Cores.append(Processor_Core.text)
    
    ROM = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table/tbody/tr[1]/td[2]/ul/li")
    Storage.append(ROM.text)
    
    RAM_tag = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table/tbody/tr[2]/td[2]/ul/li")
    RAM.append(RAM_tag.text)
    
    P_Camera = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table/tbody/tr[2]/td[2]/ul/li")
    Primary_Camera.append(P_Camera.text)
    
    S_Camera = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table/tbody/tr[5]/td[2]/ul/li")
    Secondary_Camera.append(S_Camera.text)
    
    Battery = driver.find_element_by_xpath("//div[@class='_3k-BhJ'][10]/table/tbody/tr[1]/td[2]/ul/li")
    Battery_Capacity.append(Battery.text)
    
    time.sleep(4)
```

```python
print(len(Brand_Name), len(Smartphone_Name), len(Colour), len(RAM), len(Storage), len(Primary_Camera), len(Secondary_Camera), len(Display_Size), len(Display_Resolution), len(Processor), len(Processor_Cores), len(Battery_Capacity), len(Price))
```

```python
Smartphones = pd.DataFrame({})

Smartphones['Brand Name'] = Brand_Name
Smartphones['Title'] = Smartphone_Name
Smartphones['Color'] = Colour
Smartphones['Storage'] = Storage
Smartphones['RAM'] = RAM
Smartphones['Primary Camera'] = Primary_Camera
Smartphones['Secondary Camera'] = Secondary_Camera
Smartphones['Display Size'] = Display_Size
Smartphones['Display Resolution'] = Display_Resolution
Smartphones['Processor'] = Processor
Smartphones['Processor Cores'] = Processor_Cores
Smartphones['Battery Capacity'] = Battery_Capacity
Smartphones['Price'] = Price

Smartphones
```

```python
import csv

Smartphones.to_csv('Smartphones.csv', quoting=csv.QUOTE_NONNUMERIC)
```
