# Displaying any 5 Books with Book Name, Author Name, Genre and Review from BookPage

```python
#importing the required libraries

from bs4 import BeautifulSoup
import requests
```

```python
Books = requests.get('https://bookpage.com/reviews')
Books
```

```python
Books.content
```

```python
reviews = BeautifulSoup(Books.content)
print(reviews.prettify())
```

```python
name = reviews.find_all('h4', class_='italic')
name
```

```python
book_name = []

for i in name[0:5]:
    book_name.append(i.text.strip())
    
book_name
```

```python
author = reviews.find_all('p', class_='sans bold')
author
```

```python
book_author = []

for i in author[0:5]:
    book_author.append(i.text.strip())
    
book_author
```

```python
genre = reviews.find_all('p', class_='genre-links hidden-phone')
genre
```

```python
book_genre = []

for i in genre[0:5]:
    book_genre.append(i.text.replace('\n','').strip())

book_genre
```

```python
review = reviews.find_all('div', class_='bp-block article-info')
review
```

```python
book_review = []

for i in review[0:5]:
    try:
        comment = i.find_all('p')[3].text
        book_review.append(comment)
    except:
        comment = i.find('p', class_='excerpt').text.strip()
        book_review.append(comment)
        
book_review
```

```python
print(len(book_name), len(book_author), len(book_genre), len(book_review))
```

```python
import pandas as pd
Books = pd.DataFrame({})
Books['Name'] = book_name
Books['Author'] = book_author
Books['Genre'] = book_genre
Books['Reviews'] = book_review

Books
```
