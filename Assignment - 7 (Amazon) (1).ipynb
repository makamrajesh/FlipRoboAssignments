# Scraping Data of Mobile Phones Under 20,000 from Amazon.in using Beautifulsoup

```python
#importing the required libraries

from bs4 import BeautifulSoup
import requests
```

```python
Amazon = requests.get('https://www.amazon.in/s?k=mobiles+under+20000&ref=nb_sb_noss_2')
Amazon
```

```python
Amazon.content
```

```python
Phones = BeautifulSoup(Amazon.content)
print(Phones.prettify())
```

```python
name = Phones.find_all('span', class_='a-size-medium a-color-base a-text-normal')
name
```

```python
product_name = []

for i in name[0:16]:
    product_name.append(i.text)

product_name
```

```python
price = Phones.find_all('span', class_='a-price-whole')
price
```

```python
product_price = []

for i in price[0:16]:
    product_price.append(i.text)
    
product_price
```

```python
URL = Phones.find_all('div', {'data-component-type' : "s-search-result"})
URL
```

```python
product_URL = []

for i in URL[0:16]:
    product_URL.append(i.find('img',class_='s-image')['src'])

product_URL
```

```python
Rating = Phones.find_all('span', class_='a-icon-alt')
Rating
```

```python
product_rating = []

for i in Rating[0:16]:
    product_rating.append(i.text.strip())
    
product_rating
```

```python
print(len(product_name), len(product_price), len(product_URL), len(product_rating))
```

```python
import pandas as pd

Amazon_Phones = pd.DataFrame({})

Amazon_Phones['Product'] = product_name
Amazon_Phones['Price'] = product_price
Amazon_Phones['Image_URL'] = product_URL
Amazon_Phones['Rating'] = product_rating

Amazon_Phones
```
